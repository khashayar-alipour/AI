### 1) plot
- برای رسم کردن خط صاف کاربرد داره
- نقاط با خط راست بهم وصل میشن و هر نقطه دو مختصات داره


**با استفاده از پلات به 3 روش میشه رسم کشید**
- plt.plot(x, y, marker) -> just draws with dots
- plt.plot(x, y) -> just line
- plt.plot(x, y, marker) -> line + dots



**marker settings**
- (x,y, marker='o' , ms=20 , mec='r' , mfc='b')
- marker= type
- ms= marker size
- mec= marker edge color
- mfc= marker fill color


**mec = marker edge color**
- 'r' - Red
- 'g' - Green
- 'b' - Blue
- 'c' - Cyan
- 'm' - Magenta
- 'y' - Yellow
- 'k' - Black
- 'w' - White

**Marker shapes**

- 'o' Circle
- '*' Star
- '.' Point
- ',' Pixel
- 'x' X
- 'X' X (filled)
- '+' Plus
- 'P' Plus (filled)
- 's' Square
- 'D' Diamond
- 'd' Diamond (thin)
- 'p' Pentagon
- 'H' Hexagon
- 'h' Hexagon
- 'v' Triangle Down
- '^' Triangle Up
- '<' Triangle Left
- '>' Triangle Right
- '1' Tri Down
- '2' Tri Up
- '3' Tri Left
- '4' Tri Right
- '|' Vline
- '_' Hline


**just line settings**
- ls (line styel) -> - -- .- :
- linewidth -> 10 20 30
- color or c -> r , g , b , w , ...

### 2) scatter

- مثل پلات هست و برای رسم با استفاده از نقاط کاربرد داره ولی امکانات بیشتری از پلات بهمون میده

- اسکتر فقط با نقطه رسم میکنه و مثل پلات خط پیوسته نداره

**Scatter options**

- making color list for every dot

- استفاده از رنگ های شدتی  --> viridis | inferno | ...

- using color bar

- showing 3 variable in one table using color bar

- [matplotlib colormaps](#https://matplotlib.org/stable/users/explain/colors/colormaps.html)


**scatter settings**
- s (size)

- size_list

- c or color  ->  color_list

- cmap (color map)

- alpha  ->  alpha_list

- marker

- font | loc | padding | grid




````python
x= np.linspace(0,10,30)
print(x)
````

````python
y= np.sin(x)
print(y)
````

````python
plt.scatter(x, y, marker='*',color='r')
plt.show()
````

````python
x = [1,2,3,4,5]
y= [10,20,30,40,50]
color_list=['r','r','b','g', 'b']  #number of color_list elements must be equal to dots (5 dots == 5 colors)

plt.scatter(x,y,marker='*',color=color_list)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(0,10,(100,))
y = np.random.uniform(10,20,(100,))

color_list=np.random.uniform(0,100,(100,))
size_list =np.random.uniform(20,50,(100,))
alpha_list=np.random.uniform(0,1,(100,))

plt.scatter(x,y,c=color_list,s=size_list, alpha = alpha_list ,cmap='viridis' )
plt.title('cool scatter')
plt.xlabel('random x')
plt.ylabel('random y')
plt.colorbar()

plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(0,10,(100,))
y = np.random.uniform(10,20,(100,))

color_list=np.random.uniform(0,100,(100,))
size_list =np.random.uniform(20,50,(100,))
alpha_list=np.random.uniform(0,1,(100,))


title_font={  'family' : 'serif' , 'color' : 'red' , 'size' : 20}
x_font={  'family' : 'serif' , 'color' : 'green' , 'size' : 10}
y_font={  'family' : 'serif' , 'color' : 'blue' , 'size' : 10}


plt.scatter(x, y, c=color_list, s=size_list, alpha = alpha_list ,cmap='viridis' )


plt.title('cool scatter',fontdict=title_font)
plt.xlabel('random x',fontdict=x_font)
plt.ylabel('random y',fontdict=y_font)

plt.show()
````

````python
# loc | padding | grid

title_font={  'family' : 'serif' , 'color' : 'red' , 'size' : 20}
x_font={  'family' : 'serif' , 'color' : 'green' , 'size' : 10}
y_font={  'family' : 'serif' , 'color' : 'blue' , 'size' : 10}

plt.scatter(x,y,c=color_list,s=size_list, alpha = alpha_list ,cmap='viridis' )


plt.title('nemodare scatter ziba',fontdict=title_font,loc='left',pad=30)
plt.xlabel('x haye random',fontdict=x_font)
plt.ylabel('y haye random',fontdict=y_font)
plt.grid()


plt.show()
````

### 3) pie plot

- label and percentage (درصد)

````python
lab = ['apple' , 'banana','cherry','orange']

#percentage
size = [50,30,10,10]     # -> apple=50% - banana=30% - cherry=10% - orange=10%

mycolor=['green','red','blue','black']     # -> apple=green - banana=red - cherry=blue - orange=black

ex=[0,0,0,0.2]      # -> apple=0 - banana=0 - cherry=0 - orange=0.2

plt.pie(size,labels=lab,colors=mycolor,explode=ex)
plt.show()
````

### 4) bar

- Bar Chart  -->  label and number (تعداد)

````python
x = ['A','B','C','D']
y = [3 , 7 , 1 , 10]

#1---
plt.bar(x,y,color='g',width=0.7)

#2---
plt.title('cool bar chart')
plt.ylabel('population')
plt.xlabel('group')

plt.grid(axis='y')

#3---
plt.show()
````

### 5) histograms

- برای بررسی فراوانی و انواع توزیع کاربرد داره

````python
data = np.random.uniform(0,10,(100,))   # تعداد 100 عدد رندوم بین 0 و 10 انتخاب کن که توزیع همسان دارن
# شانس انتخاب اعداد رندوم بین 0 و 10 یکسانه
# اون عدد 100 هرچی بیشتر بشه از نظر آماری اعداد رندوم توزیع یکسان تری خواهند داشت

plt.hist(data, width=0.8)

plt.show()

# X axis -> data
# Y axis -> فراوانی
````

````python
data = np.random.uniform(0,10,(100000,))

plt.hist(data, width=0.8)

plt.show()
````

````python
#mean=5 , variance=1 , shape
data = np.random.normal(5,1,(1000,))

plt.hist(data, color='g',alpha=0.5)

plt.show()
````

````python
data = np.random.normal(5,1,(10000,))
plt.hist(data,color='g',alpha=0.75,bins=50)
plt.show()
````

````python
data = np.random.normal(5,1,(10000,))
plt.hist(data,color='g',alpha=0.75,bins=500)
plt.title('random data distribition')
plt.xlabel('data')
plt.ylabel('faravani')
plt.grid()
plt.show()
````
